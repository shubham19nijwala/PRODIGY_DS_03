```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
d["job"] = label.fit_transform(d["job"])
d["marital"] = label.fit_transform(d["marital"])
d["education"] = label.fit_transform(d["education"])
d["default"] = label.fit_transform(d["default"])
d["y"] = label.fit_transform(d["y"])
d["loan"] = label.fit_transform(d["loan"])
d["contact"] = label.fit_transform(d["contact"])
d["poutcome"] = label.fit_transform(d["poutcome"])
d["housing"] = label.fit_transform(d["housing"])
d["month"] = label.fit_transform(d["month"])
```

```python
d.head()
```

```python
d.drop(["pdays","previous","poutcome"],axis = 1)
d.head()
```

```python
x = d.drop(["default"],axis = 1)
```

```python
x
```

```python
y = d["default"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train_ax,x_test_ax,y_train_ax,y_test_ax = train_test_split(x,y,test_size = 0.3)
```

```python
x_train_ax.shape
```

```python
x_test_ax.shape
```

```python
y_train_ax.shape
```

```python
y_test_ax.shape
```

# Decision Tree Classifier_Model


```python
from sklearn.tree import DecisionTreeClassifier
des_tree = DecisionTreeClassifier()
```

```python
des_tree
```

```python
des_tree.fit(x_train_ax,y_train_ax)
```

```python
y_ax = des_tree.predict(x_test_ax)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
acs = accuracy_score(y_ax,y_test_ax)*100
```

```python
acs
```

```python
confu = confusion_matrix(y_ax,y_test_ax)
```
